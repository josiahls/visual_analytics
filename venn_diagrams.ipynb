```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt


lastpass=set(['User Management','Password Sharing','User Vaults','Encription','Controls/Restrictions',
             'Zero-Knowledge Security Model','2FA','App','Web','Browser Extension','Security Auditing','$4.00/User'])
keeper=set(['User Management','Password Sharing','User Vaults','Encription','Controls/Restrictions',
             'Zero-Knowledge Security Model','2FA','App','Web','Browser Extension','Private User Vault','$2.50/User'])
dashlane=set(['User Management','Password Sharing','User Vaults','Encription','Controls/Restrictions',
             'Zero-Knowledge Security Model','2FA','App','Web','Browser Extension','Security Auditing','$4.00/User'])
v=venn3(subsets=(lastpass,keeper,dashlane), set_labels = ('LastPass','KeepSecurity','Dashlane'))
for o in list(lastpass)+list(keeper)+list(dashlane):
    try:
        v.get_label_by_id(o).set_text(o)
    except KeyError as e: pass
```
